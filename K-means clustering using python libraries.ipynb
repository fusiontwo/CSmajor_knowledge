```python
from sklearn.cluster import KMeans
```

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame(columns=['x', 'y'])
```

```python
df.loc[0] = [2, 3]
df.loc[1] = [2, 11]
df.loc[2] = [2, 18]
df.loc[3] = [4, 5]
df.loc[4] = [4, 7]
df.loc[5] = [5, 3]
df.loc[6] = [5, 15]
df.loc[7] = [6, 6]
df.loc[8] = [6, 8]
df.loc[9] = [6, 9]
df.loc[10] = [7, 2]
df.loc[11] = [7, 4]
df.loc[12] = [7, 5]
df.loc[13] = [7, 17]
df.loc[14] = [7, 18]
df.loc[15] = [8, 5]
df.loc[16] = [8, 4]
df.loc[17] = [9, 10]
df.loc[18] = [9, 11]
df.loc[19] = [9, 15]
df.loc[20] = [9, 19]
df.loc[21] = [10, 5]
df.loc[22] = [10, 8]
df.loc[23] = [10, 18]
df.loc[24] = [12, 6]
df.loc[25] = [13, 5]
df.loc[26] = [14, 11]
df.loc[27] = [15, 6]
df.loc[28] = [15, 18]
df.loc[29] = [18, 12]
```

```python
df.head(30)
```

```python
sb.lmplot('x', 'y', data=df, fit_reg=False, scatter_kws={"s": 100})
plt.title('K-means Example')
plt.xlabel('x')
plt.ylabel('y')
```

```python
# 특정 위치에서 clustering을 수행할 수 있도록 무작위값을 결정해주는 K-means++ 알고리즘이 적용되어 자동으로 clustering이 수행된다.
# 무작위값을 사용하므로 실행할 때마다 clustering의 중심값이 변화한다. 
points = df.values  # dataframe의 값들을 numpy 객체로 초기화
kmeans = KMeans(n_clusters=4).fit(points)
kmeans.cluster_centers_
```

```python
# 각 데이터들이 속한 cluster를 확인한다.
kmeans.labels_
```

```python
df['cluster'] = kmeans.labels_  # 시각화를 위해 cluster라는 속성을 만들고, 값으로는 데이터의 cluster id값이 들어가도록 한다.
df.head(30)  # 위의 데이터의 경우 총 4개(0, 1, 2, 3)의 cluster가 형성된다.
```

```python
sb.lmplot('x', 'y', data=df, fit_reg=False, scatter_kws={"s": 150}, hue="cluster")  # scatter_kws로 점의 크기 150 지정, hue는 색깔 넣기
plt.title('K-means Example')
```
